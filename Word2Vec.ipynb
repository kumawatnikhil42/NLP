```python
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess # preprocess data

story=[]

for filename in os.listdir('C:\\Users\\user\\Datasets\\gameofthrones'):
    f=open(os.path.join('C:\\Users\\user\\Datasets\\gameofthrones',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story
```

```python
model = Word2Vec(
    window=10,
    min_count=2,
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('targaryen')
```
